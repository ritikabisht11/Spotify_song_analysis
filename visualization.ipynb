```python
ax = sns.countplot(x = 'artist_count',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Top 10 Artist count in Spotify playlist
sales_state = df.groupby(['artist_count'], as_index=False)['in_spotify_playlists'].sum().sort_values(by='in_spotify_playlists', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'artist_count',y= 'in_spotify_playlists')
```

```python
# Top 10 Spotify playlists According to released year
sales_state = df.groupby(['released_year'], as_index=False)['in_spotify_playlists'].sum().sort_values(by='in_spotify_playlists', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'released_year',y= 'in_spotify_playlists')
```

```python
# Top 10 Streams in Spotify Playlist
pplr_song = df.groupby(['in_spotify_playlists'], as_index=False)['streams'].sum().sort_values(by='streams', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = pplr_song, x = 'in_spotify_playlists',y= 'streams')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'released_month')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Top 10 Artist according to Acousticness
pplr_song = df.groupby(['artist(s)_name'], as_index=False)['acousticness_%'].sum().sort_values(by='acousticness_%', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = pplr_song, x = 'artist(s)_name',y= 'acousticness_%')
```

```python
#Top 10 Artist in apple playlists
pplr_song = df.groupby(['artist(s)_name'], as_index=False)['in_apple_playlists'].sum().sort_values(by='in_apple_playlists', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = pplr_song, x = 'artist(s)_name',y= 'in_apple_playlists')
```

```python
# Top 10 Artists in Spotify playlist
pplr_song = df.groupby(['artist(s)_name'], as_index=False)['in_spotify_playlists'].sum().sort_values(by='in_spotify_playlists', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = pplr_song, x = 'artist(s)_name',y= 'in_spotify_playlists')
```

```python
# Top 10 tracks on spotify_playlist
pplr_song = df.groupby(['track_name'], as_index=False)['in_spotify_playlists'].sum().sort_values(by='in_spotify_playlists', ascending=False).head(10)

sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(data = pplr_song, x = 'track_name',y= 'in_spotify_playlists')
```

```python
# Top 5 song in deezer charts
pplr_song = df.groupby(['track_name'], as_index=False)['in_deezer_charts'].sum().sort_values(by='in_deezer_charts', ascending=False).head(5)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data = pplr_song, x = 'track_name',y= 'in_deezer_charts')
```

```python
#Top 5 songs in apple charts
pplr_song = df.groupby(['track_name'], as_index=False)['in_apple_charts'].sum().sort_values(by='in_apple_charts', ascending=False).head(5)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data = pplr_song, x = 'track_name',y= 'in_apple_charts')
```
